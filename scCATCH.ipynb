```r
library(scCATCH)
```

```r
object <- readRDS("MCA_kidney.rds")
```

```r
clu_markers <- findmarkergenes(object, species = 'Mouse', match_CellMatch = TRUE, tissue = 'Kidney')
```

```r
clu_ann <- scCATCH(clu_markers$clu_markers, species = 'Mouse', tissue = 'Kidney')
```
